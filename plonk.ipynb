### Gate constrain equation

$ g_i = q_{Li}⋅a_i+q_{Ri}⋅b_i+q_{Mi}⋅(a_i⋅b_i)+q_{Ci}+q_{Oi}⋅c_i=0 $

$ g_1 = 0⋅a_1 + 0⋅b_1 + 1⋅(a_1⋅b_1) + 0 + (-1)⋅c_1 = 0 $

$ g_2 = 2⋅a_2 + 0⋅b_2 + 0⋅(a_2⋅b_2) + 0 + (-1)⋅c_2 = 0 $

$ g_3 = 0⋅a_3 + 0⋅b_3 + 1⋅(a_3⋅b_3) + 0 + (-1)⋅c_3 = 0 $

$ g_4 = 0⋅a_4 + 0⋅b_4 + 0⋅(a_4⋅b_4) + 0 + (-1)⋅c_4 = 0 $

$ g_5 = 1⋅a_5 + (-1)⋅b_5 + 0⋅(a_5⋅b_5) + 0 + (-1)⋅c_5 = 0 $

$ g_6 = 1⋅a_6 + 0⋅b_6 + 0⋅(a_6⋅b_6) + 3 + (-1)⋅c_6 = 0 $

**selector vectors**

$ g_i = [q_{Li}, q_{Ri}, q_{Mi}, q_{Ci}, q_{Oi}] $

$ g_1 = [0, 0, 1, 0, -1] $

$ g_2 = [2, 0, 0, 0, -1] $

$ g_3 = [0, 0, 1, 0, -1] $

$ g_4 = [0, 0, 0, 0, -1] $

$ g_5 = [1, -1, 0, 0, -1] $

$ g_6 = [1, 0, 0, 3, -1] $

**let x = 2 y = 3**

```python
import galois

a = [2, 4, 3, 4, 8, -28]
b = [2, 0, 3, 9, 36, 3]
c = [4, 8, 9, 36, -28, -25]

p = 73
GF = galois.GF(p)
w = GF(6)

x = GF([1, w, w**2, w**3, w**4, w**5])

y = GF([2, 4, 3, 4, 8, GF(p-28)])
A = galois.lagrange_poly(x, y)

y = GF([2, 0, 3, 9, 36, 3])
B = galois.lagrange_poly(x, y)

y = GF([4, 8, 9, 36, GF(p-28), GF(p-25)])
C = galois.lagrange_poly(x, y)

print(f"A = {A}")
print(f"B = {B}")
print(f"C = {C}")
```
